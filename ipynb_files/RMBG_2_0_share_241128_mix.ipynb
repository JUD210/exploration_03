```python
from PIL import Image
import matplotlib.pyplot as plt
import torch
from torchvision import transforms
from transformers import AutoModelForImageSegmentation

model = AutoModelForImageSegmentation.from_pretrained('briaai/RMBG-2.0', trust_remote_code=True)
torch.set_float32_matmul_precision(['high', 'highest'][0])
model.to('cuda')
model.eval()
```

```python
def transform_image(image):
    # Convert RGBA to RGB if image has alpha channel
    if image.mode == 'RGBA':
        image = image.convert('RGB')

    image_size = (1024, 1024)
    transform = transforms.Compose([
        transforms.Resize(image_size),
        transforms.ToTensor(),
        transforms.Normalize(
            mean=[0.485, 0.456, 0.406],
            std=[0.229, 0.224, 0.225]
        )
    ])
    return transform(image)

# Usage
input_image_path = "/content/ex_image/cabbage_grandma.png"
image = Image.open(input_image_path)
print(image.size)
input_images = transform_image(image).unsqueeze(0).to('cuda')
```

```python
# Prediction
with torch.no_grad():
    preds = model(input_images)[-1].sigmoid().cpu()
pred = preds[0].squeeze()
pred_pil = transforms.ToPILImage()(pred)
mask = pred_pil.resize(image.size)
image.putalpha(mask)

image.save("no_bg_image.png")
```

```python
import cv2
import matplotlib.pyplot as plt

img_orig = cv2.imread(input_image_path)

print(img_orig.shape)

plt.imshow(cv2.cvtColor(img_orig, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
img_blurred = cv2.blur(img_orig, (13, 13))

plt.imshow(cv2.cvtColor(img_blurred, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
import numpy as np

np_mask = np.array(mask)

img_mask_color = cv2.cvtColor(np_mask, cv2.COLOR_GRAY2BGR)

img_bg_mask = cv2.bitwise_not(img_mask_color)

img_bg_blur = cv2.bitwise_and(img_blurred, img_bg_mask)
plt.imshow(cv2.cvtColor(img_bg_blur, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# 배경 이미지 경로 설정
background_image_path = "/content/ex_image/background.png"

# 배경 이미지 로드
background = Image.open(background_image_path)

# 배경 이미지를 원본 이미지 크기와 동일하게 조정하고 RGBA 모드로 변환
background = background.resize(image.size).convert("RGBA")

# 전경 이미지 설정 (이미 알파 채널이 추가된 이미지)
foreground = image  # 이전 단계에서 'image.putalpha(mask)'를 통해 알파 채널이 추가되었습니다.

# 마스크를 사용하여 전경과 배경을 합성
composite_image = Image.composite(foreground, background, mask)

# 최종 합성된 이미지 저장
composite_image.save("combined_image.png")

# 결과 이미지 표시 (Jupyter Notebook이나 Colab에서 사용 가능)
composite_image.show()
```

```python
# 새로운 입력 이미지 경로 설정
new_input_image_path = "/content/ex_image/cute_dog.png"

# 새로운 이미지 로드
new_image = Image.open(new_input_image_path)
print(f"원본 이미지 크기: {new_image.size}")

# 이미지 변환 및 전처리
new_input_images = transform_image(new_image).unsqueeze(0).to('cuda')

# 모델을 사용하여 마스크 예측
with torch.no_grad():
    new_preds = model(new_input_images)[-1].sigmoid().cpu()
new_pred = new_preds[0].squeeze()

# 예측 결과를 PIL 이미지로 변환
new_pred_pil = transforms.ToPILImage()(new_pred)

# 마스크를 원본 이미지 크기에 맞게 리사이즈
new_mask = new_pred_pil.resize(new_image.size)

# 원본 이미지에 알파 채널 추가 (마스크 적용)
new_image.putalpha(new_mask)

# 배경 이미지 로드 및 크기 조정
background_image_path = "/content/ex_image/background.png"
background = Image.open(background_image_path)
background = background.resize(new_image.size).convert("RGBA")

# 전경 이미지 설정 (알파 채널이 추가된 새로운 이미지)
foreground = new_image  # 알파 채널이 포함된 이미지

# 마스크를 사용하여 전경과 배경 합성
composite_image = Image.composite(foreground, background, new_mask)

# 합성된 이미지 저장
composite_image.save("combined_image_new.png")

# 합성된 이미지 출력
from IPython.display import display

display(composite_image)
```
